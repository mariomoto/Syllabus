## Introducción

En esta ayudantía partiremos viendo el concepto de **herencia** en OOP. La herencia es una forma de reciclar código en la definición de clases. Si queremos crear una clase nueva que se parece a una que ya tenemos y se le agregan funcionalidades, podemos hacer que la nueva clase herede de la ya creada para que mantenga todos sus métodos y atributos. En los ejercicios veremos ejemplos de esto.

## Ejercicio 1: Python League I

Empezó agosto y junto con él arrancó la competencia de fútbol pythonesco más prestigiosa del planeta, la Python League! Es por esto que te contrataron a ti, como programador experto, para programar el funcionamiento de los jugadores. Inicialmente creas una clase representando a un jugador, que contiene funcionalidades básicas. Sin embargo, te das cuenta que no sirve mucho, ya que hay ciertas cosas que solo hace un tipo de jugador. Los arqueros atajan, los defensas se barren, los delanteros tiran al arco... y justo cuando estás a punto de verter una lata de score morada sobre tu computador por la frustración, recuerdas el contenido de la semana cuatro de Programación Avanzada... **¡Herencia!**

La clase `Jugador` está implementada, y representa a un jugador cualquiera. Deberás implementar las siguientes clases:

- `Arquero`: representa a un arquero. Tiene los siguientes métodos:
  - `__init__`: pide los mismos argumentos que el inicializador de `Jugador` y además un argumento `porc_atajadas`, un float entre 0 y 1 que indica el porcentaje de balones que ataja. La función debe llamar al inicializador de `Jugador`.
  - `atajar`: primero, el método marca el atributo `tiene_balon` como `True`. Luego, el arquero atajará el balón con una probabilidad dada por su porcentaje de atajadas. Si lo ataja, imprimirá `"Atajadón de {apellido}!!! Hay portero señores!!!!"` y retornará `True`, en otro caso retorna `False`.

- `Defensa`: representa a un defensa. Tiene los siguientes métodos:
  - `__init__`: pide los mismos argumentos que el inicializador de `Jugador` y además un argumento `porc_barridas`, un float entre 0 y 1 que indica el porcentaje de barridas exitosas. La función debe llamar al inicializador de `Jugador`.
  - `barrerse`: pide como argumento el `rival` al cual el defensa le quiere hacer la barrida. El defensa se barrerá exitosamente con una probabilidad `porc_baridas`. Si lo logra, imprimirá `"{apellido} le quita el balón a {apellido_rival}"`, marcará el atributo `tiene_balon` del rival como `False`y el propio como `True`. Si no lo logra, imprimirá `"{apellido} le trata de quitar el balón a {apellido_rival}... pero no puede!"`

- `Delantero`: representa a un delantero. Tiene los siguientes métodos:
  - `__init__`: pide los mismos argumentos que el inicializador de `Jugador` y además un argumento `porc_tiros`, un float entre 0 y 1 que indica el porcentaje de tiros exitosos. La función debe llamar al inicializador de `Jugador`.
  - `tirar`: pide como argumento el `arquero_rival`. El jugador chuteará al arco con una probabilidad `porc_tiros`. De ser un tiro exitoso, imprimirá `"{apellido} chuta..."`, y luego llamará al método `atajar` del `arquero_rival`, si este retorna `False`, imprimirá `"GOLAZO DEL {equipo}!!! Un tiro al ángulo de {apellido}!!"`. Si el tiro no es exitoso, imprimirá `"{apellido} chuta... y fuera!"`, cambiará el atributo `tiene_balon` del `arquero_rival` a `True`, y el propio a `False`.

Hint: si quieres que algo pase con cierta probabilidad, puedes usar la función `random` del módulo `random`, que te entrega un float al azar entre 0 y 1. Piensa, por ejemplo, cuál es la probabilidad de que un número al azar entre 0 y 1 sea menor a 0.7 🤔.

```python
from random import random


class Jugador:

    def __init__(self, apellido, numero, equipo):
        self.apellido = apellido
        self.numero = numero
        self.equipo = equipo
        self.tiene_balon = False
    
    def pasar(self, jugador_2):
        print(f"{self.apellido} se la pasa a {jugador_2.apellido}")
        self.tiene_balon = False
        jugador_2.tiene_balon = True

class Arquero(Jugador):

    def __init__(self, apellido, numero, equipo, porc_atajadas):
        super().__init__(apellido, numero, equipo)
        self.porc_atajadas = porc_atajadas

    def atajar(self) -> bool:
        apellido = self.apellido
        self.tiene_balon = True
        if random() < self.porc_atajadas:
            print(f"Atajadón de {apellido}!!! Hay portero señores!!!!")
            return True
        else:
            return False

class Defensa(Jugador):

    def __init__(self, apellido, numero, equipo, porc_barridas):
        super().__init__(apellido, numero, equipo)
        self.porc_barridas = porc_barridas

    def barrerse(self, rival: Jugador):
        apellido = self.apellido
        apellido_rival = rival.apellido
        if random() < self.porc_barridas:
            print(f"{apellido} le quita el balón a {apellido_rival}")
            rival.tiene_balon = False
            self.tiene_balon = True
        else:
            print(f"{apellido} le trata de quitar el balón a {apellido_rival}... pero no puede!")

class Delantero(Jugador):

    def __init__(self, apellido, numero, equipo, porc_tiros):
        super().__init__(apellido, numero, equipo)
        self.porc_tiros = porc_tiros

    def tirar(self, arquero_rival: Arquero):
        apellido = self.apellido
        equipo = self.equipo
        if random() < self.porc_tiros:
            print(f"{apellido} chuta...")
            if not arquero_rival.atajar():
                print(f"GOLAZO DEL {equipo}!!! Un tiro al ángulo de {apellido}!!")
        else:
            print(f"{apellido} chuta... y fuera!")
            arquero_rival.tiene_balon = True
            self.tiene_balon = False

listerpool = [Arquero("Becker", 1, "Listerpool", 0.55),
              Defensa("Van Dijk", 4, "Listerpool", 0.55),
              Defensa("Konaté", 5, "Listerpool", 0.40),
              Delantero("Salah", 11, "Listerpool", 0.95),
              Delantero("Ekitike", 22, "Listerpool", 0.8)]

mapchester_city = [Arquero("Trafford", 1, "Mapchester City", 0.3),
                   Defensa("Stones", 5, "Mapchester City", 0.45),
                   Defensa("Dias", 3, "Mapchester City", 0.35),
                   Delantero("Haaland", 9, "Mapchester City", 0.95),
                   Delantero("Marmoush", 7, "Mapchester City", 0.8)]

listerpool[0].tiene_balon = True
listerpool[0].pasar(listerpool[1])
listerpool[1].pasar(listerpool[4])
mapchester_city[1].barrerse(listerpool[4])
if mapchester_city[1].tiene_balon:
    mapchester_city[1].pasar(mapchester_city[4])
    listerpool[2].barrerse(mapchester_city[4])
    if mapchester_city[4].tiene_balon:
        mapchester_city[4].pasar(mapchester_city[3])
        mapchester_city[3].tirar(listerpool[0])
else:
    listerpool[4].pasar(listerpool[3])
    listerpool[3].tirar(mapchester_city[0])
```

## Ejercicio 2: Python League II

Primero, harás las siguientes modificaciones a la clase `Delantero`:

Un cabezazo fulminante de {apellido}!!"` en caso de meter un gol y `"{apellido} cabecea... y fuera!"`  en caso de cabecear hacia afuera.

Luego implementarás la clase `Lateral`, que hereda de `Defensa` y `Delantero`. Tiene los siguientes métodos:

- `__init__`: pide todos los argumentos que necesita un `Defensa` y un `Delantero`, y además un argumento `porc_centros`, un float entre 0 y 1 que indica el porcentaje de centros exitosos. La función debe hacer solo **una** llamada a `super`.

```python
class Defensa(Jugador):

    def __init__(self, apellido, numero, equipo, porc_barridas, **kwargs):
        super().__init__(apellido, numero, equipo, **kwargs)
        self.porc_barridas = porc_barridas

    def barrerse(self, rival: Jugador):
        apellido = self.apellido
        apellido_rival = rival.apellido
        if random() < self.porc_barridas:
            print(f"{apellido} le quita el balón a {apellido_rival}")
            rival.tiene_balon = False
            self.tiene_balon = True
        else:
            print(f"{apellido} le trata de quitar el balón a {apellido_rival}... pero no puede!")

class Delantero(Jugador):

    def __init__(self, apellido, numero, equipo, porc_tiros, porc_cabeceo, **kwargs):
        super().__init__(apellido, numero, equipo, **kwargs)
        self.porc_tiros = porc_tiros
        self.porc_cabeceo = porc_cabeceo

    def tirar(self, arquero_rival: Arquero):
        apellido = self.apellido
        equipo = self.equipo
        if random() < self.porc_tiros:
            print(f"{apellido} chuta...")
            if not arquero_rival.atajar():
                print(f"GOLAZO DEL {equipo}!!! Un tiro al ángulo de {apellido}!!")
        else:
            print(f"{apellido} chuta... y fuera!")
            arquero_rival.tiene_balon = True
            self.tiene_balon = False

    def cabecear(self, arquero_rival: Arquero):
        apellido = self.apellido
        equipo = self.equipo
        if random() < self.porc_cabeceo:
            print(f"{apellido} cabecea...")
            if not arquero_rival.atajar():
                print(f"GOLAZO DEL {equipo}!!! Un cabezazo fulminante de {apellido}!!")
        else:
            print(f"{apellido} cabecea... y fuera!")
            arquero_rival.tiene_balon = True
            self.tiene_balon = False

class Lateral(Defensa, Delantero):

    def __init__(self, apellido, numero, equipo, porc_centros,  **kwargs):
        super().__init__(apellido, numero, equipo,  **kwargs)
        self.porc_centros = porc_centros

    def centrar(self, delantero: Delantero, arquero_rival: Arquero):
        apellido = self.apellido
        apellido_delantero = delantero.apellido
        if random() < self.porc_centros:
            print(f"Centro de {apellido} para {apellido_delantero}")
            delantero.tiene_balon = True
            self.tiene_balon = False
            delantero.cabecear(arquero_rival)
        else:
            print(f"Centro de {apellido}... se va fuera!")
            arquero_rival.tiene_balon = True
            self.tiene_balon = False

listerpool = [Arquero(apellido="Becker", numero=1, equipo="Listerpool", porc_atajadas=0.55),
              Defensa(apellido="Van Dijk", numero=4, equipo="Listerpool", porc_barridas=0.55),
              Defensa(apellido="Konaté", numero=5, equipo="Listerpool", porc_barridas=0.4),
              Lateral(apellido="Kerkez", numero=6, equipo="Listerpool", porc_barridas=0.4, porc_cabeceo=0.6, porc_tiros=0.7, porc_centros=0.75),
              Delantero(apellido="Salah", numero=11, equipo="Listerpool", porc_tiros=0.95, porc_cabeceo=0.8),
              Delantero(apellido="Ekitike", numero=22, equipo="Listerpool", porc_tiros=0.8, porc_cabeceo=0.8)]

mapchester_city = [Arquero(apellido="Trafford", numero=1, equipo="Mapchester City", porc_atajadas=0.3),
                   Defensa(apellido="Stones", numero=5, equipo="Mapchester City", porc_barridas=0.45),
                   Defensa(apellido="Dias", numero=3, equipo="Mapchester City", porc_barridas=0.35),
                   Lateral(apellido="Gvardiol", numero=24, equipo="Mapchester City", porc_barridas=0.3, porc_cabeceo=0.5, porc_tiros=0.8, porc_centros=0.6),
                   Delantero(apellido="Haaland", numero=9, equipo="Mapchester City", porc_tiros=0.95, porc_cabeceo=0.99),
                   Delantero(apellido="Marmoush", numero=7, equipo="Mapchester City", porc_tiros=0.8, porc_cabeceo=0.9)]

listerpool[0].tiene_balon = True
listerpool[0].pasar(listerpool[1])
listerpool[1].pasar(listerpool[3])
mapchester_city[3].barrerse(listerpool[3])
if mapchester_city[3].tiene_balon:
    listerpool[2].barrerse(mapchester_city[3])
    if mapchester_city[3].tiene_balon:
        mapchester_city[3].tirar(listerpool[0])
else:
    listerpool[3].centrar(listerpool[5], mapchester_city[0])
```

## Ejercicio 3: Python League III

En este ejercicio usaremos **properties**. Las **properties** son funcionalidades en python que nos permiten, en una clase, tener datos que se acceden como atributos pero por detrás funcionan como métodos. Una **property** siempre debe tener un **getter**, que sirve para acceder a ella. Adicionalmente, pueden tener un **setter**, que sirve para modificarla.

Finalmente lograste descifrar las tácticas de tu colega pelado, y vas como avión. Sin embargo, notas que hay un problema. Cuando los delanteros no logran hacer un tiro exitoso, los porteros se ven envalentonados por esto, y su porcentaje de atajadas sube por **0.05**. Esto lleva a que algunos porteros tengan un porcentaje de atajadas muy alto, e incluso mayor a uno. Esto te parece inconcebible, y te determinas a arreglarlo usando los últimos contenidos que aprendiste en Programación Avanzada: **properties**.

Deberás añadirle a la clase `Arquero` una **property** llamada `porc_atajadas`, que se comporta de la siguiente manera:
- getter: imprimirás `"Accediendo a la property..."`, para mostrar que estás accediendo a la función. Retornarás el valor del atributo asociado.
- setter: recibirás un nuevo valor para el porcentaje de atajadas. Si este valor es menor o igual a **0.7**, cambiarás el valor del atributo asociado redondeando a 2 decimales e imprimirás `"Debido a ese tiro tan malo, {apellido} sube su porcentaje de atajadas a {nuevo_valor}"`. Si el valor es mayor a **0.7**, imprimirás `"{self.apellido} no puede subir más su porcentaje de atajadas!"`.

También deberás modificar el siguiente método de la clase `Delantero`:
- `tirar`: si el tiro no es exitoso, deberás añadirle **0.05** al porcentaje de atajadas del `arquero_rival`, usando la **property**. 

Ojo: para que el ejercicio funcione bien, debes copiar de nuevo las definiciones anteriores de `Defensa` y `Lateral`.

```python
class Arquero(Jugador):

    def __init__(self, apellido, numero, equipo, porc_atajadas):
        super().__init__(apellido, numero, equipo)
        self._porc_atajadas = porc_atajadas

    @property
    def porc_atajadas(self) -> float:
        print("Accediendo a la property...")
        return self._porc_atajadas
    
    @porc_atajadas.setter
    def porc_atajadas(self, nuevo_valor):
        apellido = self.apellido
        nuevo_valor = round(nuevo_valor, 2)
        if nuevo_valor <= 0.7:
            print(f"Debido a ese tiro tan malo, {apellido} sube su porcentaje de atajadas a {nuevo_valor}")
            self._porc_atajadas = nuevo_valor
        else:
            print(f"{apellido} no puede subir más su porcentaje de atajadas!")
    
    def atajar(self) -> bool:
        apellido = self.apellido
        self.tiene_balon = True
        if random() < self._porc_atajadas:
            print(f"Atajadón de {apellido}!!! Hay portero señores!!!!")
            return True
        else:
            return False

class Defensa(Jugador):

    def __init__(self, apellido, numero, equipo, porc_barridas, **kwargs):
        super().__init__(apellido, numero, equipo, **kwargs)
        self.porc_barridas = porc_barridas

    def barrerse(self, rival: Jugador):
        apellido = self.apellido
        apellido_rival = rival.apellido
        if random() < self.porc_barridas:
            print(f"{apellido} le quita el balón a {apellido_rival}")
            rival.tiene_balon = False
            self.tiene_balon = True
        else:
            print(f"{apellido} le trata de quitar el balón a {apellido_rival}... pero no puede!")

class Delantero(Jugador):

    def __init__(self, apellido, numero, equipo, porc_tiros, porc_cabeceo, **kwargs):
        super().__init__(apellido, numero, equipo, **kwargs)
        self.porc_tiros = porc_tiros
        self.porc_cabeceo = porc_cabeceo

    def tirar(self, arquero_rival: Arquero):
        apellido = self.apellido
        equipo = self.equipo
        if random() < self.porc_tiros:
            print(f"{apellido} chuta...")
            if not arquero_rival.atajar():
                print(f"GOLAZO DEL {equipo}!!! Un tiro al ángulo de {apellido}!!")
        else:
            print(f"{apellido} chuta... y fuera!")
            arquero_rival.tiene_balon = True
            arquero_rival.porc_atajadas += 0.05
            self.tiene_balon = False

    def cabecear(self, arquero_rival: Arquero):
        apellido = self.apellido
        equipo = self.equipo
        if random() < self.porc_cabeceo:
            print(f"{apellido} cabecea...")
            if not arquero_rival.atajar():
                print(f"GOLAZO DEL {equipo}!!! Un cabezazo fulminante de {apellido}!!")
        else:
            print(f"{apellido} cabecea... y fuera!")
            arquero_rival.tiene_balon = True
            self.tiene_balon = False

class Lateral(Defensa, Delantero):

    def __init__(self, apellido, numero, equipo, porc_centros,  **kwargs):
        super().__init__(apellido, numero, equipo,  **kwargs)
        self.porc_centros = porc_centros

    def centrar(self, delantero: Delantero, arquero_rival: Arquero):
        apellido = self.apellido
        apellido_delantero = delantero.apellido
        if random() < self.porc_centros:
            print(f"Centro de {apellido} para {apellido_delantero}")
            delantero.tiene_balon = True
            self.tiene_balon = False
            delantero.cabecear(arquero_rival)
        else:
            print(f"Centro de {apellido}... se va fuera!")
            arquero_rival.tiene_balon = True
            self.tiene_balon = False

listerpool = [Arquero(apellido="Becker", numero=1, equipo="Listerpool", porc_atajadas=0.55),
              Defensa(apellido="Van Dijk", numero=4, equipo="Listerpool", porc_barridas=0.55),
              Defensa(apellido="Konaté", numero=5, equipo="Listerpool", porc_barridas=0.4),
              Lateral(apellido="Kerkez", numero=6, equipo="Listerpool", porc_barridas=0.4, porc_cabeceo=0.6, porc_tiros=0.7, porc_centros=0.75),
              Delantero(apellido="Salah", numero=11, equipo="Listerpool", porc_tiros=0.95, porc_cabeceo=0.8),
              Delantero(apellido="Ekitike", numero=22, equipo="Listerpool", porc_tiros=0.8, porc_cabeceo=0.8)]

mapchester_city = [Arquero(apellido="Trafford", numero=1, equipo="Mapchester City", porc_atajadas=0.3),
                   Defensa(apellido="Stones", numero=5, equipo="Mapchester City", porc_barridas=0.45),
                   Defensa(apellido="Dias", numero=3, equipo="Mapchester City", porc_barridas=0.35),
                   Lateral(apellido="Gvardiol", numero=24, equipo="Mapchester City", porc_barridas=0.3, porc_cabeceo=0.5, porc_tiros=0.3, porc_centros=0.6),
                   Delantero(apellido="Haaland", numero=9, equipo="Mapchester City", porc_tiros=0.95, porc_cabeceo=0.99),
                   Delantero(apellido="Marmoush", numero=7, equipo="Mapchester City", porc_tiros=0.8, porc_cabeceo=0.9)]


mapchester_city[0].tiene_balon = True
mapchester_city[0].pasar(mapchester_city[3])
mapchester_city[3].tirar(listerpool[0])
listerpool[0].tiene_balon = False
mapchester_city[0].tiene_balon = True
mapchester_city[0].pasar(mapchester_city[3])
mapchester_city[3].tirar(listerpool[0])
listerpool[0].tiene_balon = False
mapchester_city[0].tiene_balon = True
mapchester_city[0].pasar(mapchester_city[3])
mapchester_city[3].tirar(listerpool[0])
listerpool[0].tiene_balon = False
mapchester_city[0].tiene_balon = True
mapchester_city[0].pasar(mapchester_city[3])
mapchester_city[3].tirar(listerpool[0])
listerpool[0].tiene_balon = False
mapchester_city[0].tiene_balon = True
mapchester_city[0].pasar(mapchester_city[3])
mapchester_city[3].tirar(listerpool[0])
print(listerpool[0].porc_atajadas)
```
